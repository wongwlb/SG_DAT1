## 10. Display a histogram of ages of all killings

```python
import matplotlib.pyplot as plt
%matplotlib inline

killings.age.plot(kind='hist', bins=10)
plt.xlabel('Age')
plt.ylabel('No. of Killings')
plt.title('Histogram of Killings by Age')
```

## 13. Show a bar chart with Counts of Deaths every Month

```python
xmonth = killings.month.value_counts()

xmonth.ix[['January', 'February', 'March', 'April', 'May', 'June']].plot(kind='bar')
# ix chronologically orders the months Jan - June, thanks Ilya for the note
plt.xlabel('Month')
plt.ylabel('Counts of Death')
plt.title('Bar Plot of Counts of Deaths every Month')
```
